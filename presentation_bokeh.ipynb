```python
from bokeh.plotting import figure, show

p = figure(title="Basic Line Plot")
p.line([1, 2, 3], [4, 6, 5], line_width=2)

show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Create CDS
source = ColumnDataSource(data={
    "x": [1, 2, 3, 4],
    "y": [4, 7, 1, 6]
})

# Plot using CDS
p = figure(title="ColumnDataSource Example")
p.circle("x", "y", size=10, source=source, color="navy")

show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

p = figure(
    title="Tools Example",
    tools="pan,wheel_zoom,box_zoom,reset,hover"
)

p.circle([1, 2, 3, 4], [5, 6, 2, 7], size=15)

# Customize hover tool
hover = p.select_one(HoverTool)
hover.tooltips = [
    ("x", "$x"),
    ("y", "$y"),
]

show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.layouts import row, column

# Two simple plots
p1 = figure(title="Plot 1")
p1.line([1,2,3], [4,5,3], line_width=2)

p2 = figure(title="Plot 2")
p2.circle([1,2,3], [3,6,2], size=10, color="red")

# Layout
layout = row(p1, p2)
# or: layout = column(p1, p2)

show(layout)
```

```python
from bokeh.layouts import column
from bokeh.models import Slider
from bokeh.plotting import figure, show
from bokeh.io import curdoc

p = figure()
line = p.line([1,2,3], [1,4,9])

slider = Slider(start=1, end=10, value=1, title="Multiplier")

def update(attr, old, new):
    line.data_source.data["y"] = [i * slider.value for i in [1,4,9]]

slider.on_change("value", update)

curdoc().add_root(column(slider, p))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
days = [1,2,3,4,5,6,7]
views = [100,500,3000,8000,15000,12000,18000]

#create plot
plt.plot (days,views)

#add titles and labels
plt.title("Tiktok Videos views Over Time")
plt.xlabel("Days Since Posting")
plt.ylabel("Total Views")

#display plot
plt.show()
print("Basic line plot created!")
```

```python
#comparing performance of different social media posts 

#data for three different tiktok videos
days = [1,2,3,4,5,6,7]
dailouge_video_views = [100,500,3000,8000,15000,12000,18000]
cooking_video_views = [50,200,1000,3000,8000,15000,25000]
prank_video_views = [200,800,1500,4000,6000,5000,7000]

plt.plot(days,dailouge_video_views,label = 'Dailouge Video')
plt.plot(days,cooking_video_views,label = 'Cooking Video')
plt.plot(days,prank_video_views,label = 'Prank Video')

plt.title("TikTok Videos Performance Comparison")
plt.xlabel("Days since Posting")
plt.ylabel("Total Views")

plt.legend()
plt.grid(True,alpha = 0.3)

plt.show()
```

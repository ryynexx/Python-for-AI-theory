```python
students = pd.DataFrame({
    'ID': [1,2,3],
    'Name': ['Ali','Sara','Zain']
})

marks = pd.DataFrame({
    'ID': [1,2,3],
    'Score': [85, 90, 78]
})
merged_df = students.merge(marks, on='ID')
print(merged_df)
```

```python
concat_vert = pd.concat([students, students], ignore_index=True)
print(concat_vert)
```

```python
concat_horiz = pd.concat([students, marks], axis=1)
print(concat_horiz)
```

```python
left_join = students.merge(marks, on='ID', how='left')
print(left_join)
```

```python
extra = pd.DataFrame({'ID':[1,2,3],'Age':[20,21,22]})
merged_three = students.merge(marks, on='ID').merge(extra, on='ID')
print(merged_three)
```
